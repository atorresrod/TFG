En este cuaderno de jupyter se crea el histograma utilizado para visualizar y comparar los resultados del agente DQN implementado en este TFG frente a los resultados obtenidos por Mnih et al (2015).

```python
import matplotlib.pyplot as plt
import numpy as np

envs = ['Breakout', 'Space Invaders', 'Enduro', 'Video Pinball']
dqn_rewards = [195.5, 1349, 545.6, 191855.8]
dqn_std = [117.1, 346.9, 748, 206344.6]
mnih_rewards = [401.2, 1976, 301.8, 42684]
std_mnih = [26.9, 893, 24.6, 16287]

bar_width = 0.35

# Bar position
r1 = np.arange(len(envs))
r2 = [x + bar_width for x in r1]

plt.bar(r1, dqn_rewards, yerr=dqn_std, width=bar_width, label='DQN', capsize=5)
plt.bar(r2, mnih_rewards, yerr=std_mnih, width=bar_width, label='Mnih et al. 2015', capsize=5)

plt.xlabel('Entornos')
plt.ylabel('Recompensa media')
plt.title('Comparación de recompensas medias por entorno')
plt.xticks([r + bar_width/2 for r in range(len(envs))], envs)
plt.yscale('log')
plt.legend()
plt.show()
```
