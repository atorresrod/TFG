El propósito de este notebook es obtener imágenes de los entornos utilizados para usarlas en la memoria del TFG

```python
import gymnasium as gym
import matplotlib.pyplot as plt
```

```python
env = gym.make("FrozenLake-v1", map_name="8x8", render_mode="rgb_array")

env.reset()
env.step(0)

img = env.render()

plt.axis('off')
_ = plt.imshow(img)
```
